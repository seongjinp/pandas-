```python
#xlsx 모듈 연습

#모듈 불러오기
from openpyxl import *

#워크북 생성
wb = Workbook()

#시트 활성화

ws = wb.active

#시트 추가

ws2 = wb.create_sheet()

#시트 이름 정하기

ws.title = '1번 시트'
ws2.title = '2번 시트'

#셀에 값 넣기

ws['A1'] = '박성진'
ws['A2'] = '조성지'
ws['B1'] = '전소연'
ws['B2'] = '진윤정'

#출력하기
a = ws.iter_rows(min_row=1,max_row=2,min_col=1,max_col=2) #generator
b = ws.iter_cols(min_col=1,max_col=2,min_row=1,max_row=2) #generator

for i in a :
    for j in i :
        print(j.value)

print()

for i in b :
    for j in i :
        print(j.value)

print()
a = ws['A1':'B2'] #튜플

for i in a :
    for j in i :
        print(j.value)

# 반복하여 입력하기

dic1 = {'이름':'박성진','나이':30,'사는 곳':'대전'}
dic2 = {'이름':'조성지','나이':29,'사는 곳':'공주'}
dic3 = {'이름':'진윤정','나이':29,'사는 곳':'대구'}
dic4 = {'이름':'김두한','나이':38,'사는 곳':'평양'}
list1 = [dic1,dic2,dic3,dic4]

# 1. title 만들기
i = 1
for key in dic1.keys() :
    ws2[f'{chr(64+i)}'+str(1)] = key
    i += 1
    
# 2. value 넣기
j = 1
for dic in list1 :
    i = 1
    for value in dic.values() :
        ws2[f'{chr(64+i)}'+str(j+1)] = value
        i += 1
    j += 1



# 전체 출력하기
print()

for sheet in wb.worksheets :
    print('sheet name = {}'.format(sheet.title))
    for row in sheet.rows : #시트를 한줄씩(row,행) 읽음 a1,b1 // a2,b2 이런식
        for idx,cell in enumerate(row) : # 행의 인덱스와 벨류값의 generator
            if idx != 0 :
                print('\t',end='')
            if cell.value == None :
                continue
            print(cell.value,end='')
        print()
    print()

# 저장하기
wb.save('엑셀연습.xlsx')

```
