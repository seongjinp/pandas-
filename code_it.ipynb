```python
import numpy as np
import pandas as pd

df = pd.read_csv('pandas/world_cities.csv')
df.head(5)

df['Country'].describe()
```

```python
df['City / Urban area'].describe()
```

```python
condition = (df['Population']/df['Land area (in sqKm)']) > 10000
df[condition]['City / Urban area'].info()
```

```python
density = df['Population']/df['Land area (in sqKm)']
condition = density == density.max()
df[condition]['City / Urban area']
```

```python
# 시각화

import numpy as np
%matplotlib inline
import pandas as pd

df = pd.read_csv('broadcast.csv', index_col=0)

df.plot(kind='line') # 선그래프는 생략해도 됨(kind='line')
```

```python
#한가지 방송사에 대한 그래프만

df.plot(y='KBS')
```

```python
df.plot(y=['KBS','JTBC'])
```

```python
# 막대그래프

df = pd.read_csv('sports.csv', index_col=0)

df.plot(kind='bar')
```

```python
df.plot(kind='barh') # 가로막대 그래프 (bar horizental)
```

```python
df.plot(kind='bar', stacked=True) # 쌓아서 출력
```
