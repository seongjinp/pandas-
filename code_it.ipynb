```python
df = pd.read_csv('titanic.csv')

# 1. 타이타닉의 승객은 30대와 40대가 가장 많다.
df['나이대'] = (df['Age']//10)*10
df['나이대'].value_counts().sort_values(ascending=False)
df.plot(kind='hist', y='Age', bins=30)
# -> 20대가 가장많고 그다음 30대

# 2. 가장 높은 요금을 낸사람은 30대이다. (맞음)
df[['Age','Fare']].plot('Age','Fare',kind='scatter')

# 3. 생존자가 사망자보다 더 많다. (0사망, 1생존)
df['Survived'].value_counts()
# -> 사망자가 549명으로 가장 많다.

# 4. 1등실, 2등실, 3등실 중 가장 많은 사람이 탑승한 곳은 3등실이다. (맞음)
df['Pclass'].value_counts()

# 5. 가장 생존율이 높은 객실 등급은 1등실이다. (맞음)
df.loc[df['Survived']==1,'Pclass'].value_counts()

# 6. 나이가 어릴수록 생존율이 높다.
sns.stripplot(data=df,x='Survived',y='Age')
sns.violinplot(data=df, x='Survived', y='Age')

# 7. 나이보다 성별이 생존율에 더 많은 영향을 미친다.
sns.stripplot(data=df, x="Survived", y="Age", hue="Sex")
```

```python
# 데이터 합치기 (merge)

price = pd.read_csv('vegetable_price.csv')
quantity = pd.read_csv('vegetable_quantity.csv')
# Inner Join (두 데이터프레임에서 겹치는 부분만 합침)

pd.merge(price,quantity, on='Product')
```

```python
# Left Outer Join : 왼쪽 데이터 기준으로 겹치는 부분을 합침
pd.merge(price,quantity,on='Product',how='left')
```

```python
# right outer join : 오른쪽 데이터 기준으로 겹치는 부분을 합침
pd.merge(price,quantity,on='Product',how='right')
```

```python
# Full outer join : 모두다 합침
pd.merge(price,quantity,on='Product',how='outer')
```
